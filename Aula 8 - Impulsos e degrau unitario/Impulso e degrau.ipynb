# Impulso e degrau

Neste notebook vamos falar sobre os sinais ***impulso unit치rio*** e ***degrau unit치rio***

```python
# importar as bibliotecas necess치rias
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
```

## 1. Impulso unit치rio

$\delta(t)$


```python
# Um impulso unit치rio
t = np.arange(start = -2, stop = 2, step=0.1) # vetor temporal
xt = np.zeros(len(t))
xt[np.where(np.logical_and(t>=-2e-10,t<=2e-10))] = 1 #complicado assim s칩 por conta de arredondamentos num칠ricos (as vezes t=0 n칚o 칠 parte do vetor temporal)

plt.figure()
plt.stem(t, xt, '-b', label = r'$\delta(t)$', basefmt=" ", use_line_collection=  True)
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel('Amplitude [-]')
plt.ylim((-0.5, 1.5))
plt.tight_layout()   
plt.show()
```

## 2. Impulso unit치rio deslocado

$\delta(t-t_0)$

```python
# Um impulso unit치rio
t = np.arange(start = -2, stop = 2, step=0.1)  # vetor temporal
xt = np.zeros(len(t))
xt[np.where(np.logical_and(t>=1-2e-10,t<=1+2e-10))] = 1 # 

plt.figure()
plt.stem(t, xt, '-b', label = r'$\delta(t-1)$', basefmt=" ", use_line_collection=  True)
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel('Amplitude [-]')
plt.ylim((-0.5, 1.5))
plt.tight_layout()   
plt.show()
```

## 3. Trem de impulsos

O trem de impulsos 칠 dado por:

\begin{equation}
s(t) = \sum_{n = -\infty}^{\infty}\delta(t-n\Delta)
\end{equation}
em que $\Delta$ [s] 칠 chamado de ***per칤odo de amostragem***. De fato, $\Delta = 1/F_s$, em que $F_s$ 칠 a frequ칡ncia de amostragem ou o n칰mero de amostras por segundo que a sua placa de som ou sistema de aquisi칞칚o obt칠m.

```python
# Frequ칡ncia de amostragem
Fs = 50
Delta = 1/Fs

t = np.arange(start = -2, stop = 2, step = Delta)
st = np.ones(len(t))

plt.figure()
plt.stem(t, st, '-b', label = r'$s(t)$', basefmt=" ", use_line_collection=  True)
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel('Amplitude [-]')
plt.xlim((-0.05, 0.05))
plt.ylim((-0.5, 1.5))
plt.tight_layout()   
plt.show()
```

# 4. Degrau unit치rio

Esta fun칞칚o est치 relacionada ao impulso unit치rio da seguinte forma

\begin{equation}
u(t) = \int_{-\infty}^{t}\delta(\tau) \mathrm{d} \tau
\end{equation}

ou

\begin{equation}
\frac{\mathrm{d}u(t)}{\mathrm{d} t} = \delta(t) 
\end{equation}

```python
t = np.arange(start = -2, stop = 2, step=0.01)  # vetor temporal
xt = np.ones(len(t))
xt[t<0] = 0 # 

plt.figure()
plt.plot(t, xt, '-b', label = r'$u(t)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel('Amplitude [-]')
plt.ylim((-0.5, 1.5))
plt.tight_layout()   
plt.show()
```
