```python
# importar as bibliotecas necess√°rias
import numpy as np # arrays
import matplotlib.pyplot as plt # plots
plt.rcParams.update({'font.size': 14})
import IPython.display as ipd # to play signals
import sounddevice as sd
import soundfile as sf
```

```python
##### x(t) organ signal #################
xt, fs = sf.read('373703__samulis__organ-loud-c4-rode-pedal-25.wav')
time_xt = np.linspace(0, (len(xt)-1)/fs, len(xt))
time_lim = (0, time_xt[-1])#(1, 1.03)

plt.figure(figsize=(12,3))
plt.plot(time_xt, xt[:,0], '-b', linewidth = 1, label = r'$x_L(t)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude [-]')
plt.xlim(time_lim)
plt.tight_layout()
plt.show()

# x(t)
ipd.Audio(xt[:,0], rate=fs) 
```

```python
# An√°lise espectral
XwL = np.fft.fft(xt[:,0])
freq = np.linspace(0, (len(xt)-1)*fs/len(xt), len(xt))

plt.figure()
plt.semilogx(freq, 20*np.log10(np.abs(XwL)), '-b', linewidth = 1, label = r'$X(\mathrm{j} \omega)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Frequ√™ncia [Hz]')
plt.ylabel('Amplitude [-]')
plt.xlim((100,2000))
plt.ylim((50,80))
plt.tight_layout()
plt.show()
```
