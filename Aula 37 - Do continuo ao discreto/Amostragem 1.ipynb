# Do contínuo ao discreto

Neste notebook exploramos como é um sinal discreto no computador. Vamos mostrar duas versões de um seno amostrado.

```python
# importar as bibliotecas necessárias
import numpy as np # arrays
import matplotlib.pyplot as plt # plots
plt.rcParams.update({'font.size': 14})
```

## Exemplo 1 

Vamos criar um seno, entre 0 [s] e 1[s], com frequência 2 [Hz]. Neste primeiro exemplo a taxa de amostragem será alta ($F_{s1} = 1000 [Hz])$

```python
Fs_1 = 1000
time = np.arange(0, 1, 1/Fs_1)
xt_1 = np.sin(2*np.pi*600*time)

plt.figure()
#plt.plot(time, xt_1, 'r', linewidth = 2)
plt.stem(time, xt_1)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude [-]')
plt.grid(which='both', axis='both')
plt.xlim((0,1))
plt.ylim((-1.6,1.6));
```

Vamos explorar aspectos como o tamanho do sinal (número de amostras) e o vetor temporal

```python
print('O número de amostras no sinal é {}.'.format(len(time)))
print('o vetor temporal é {}'.format(time))
```

## Exemplo 2

Vamos criar um seno, entre 0 [s] e 1[s], com frequência 2 [Hz]. Neste segundo exemplo vamos mudar a taxa de amostragem e averiguar o que ocorre.

```python
Fs_2 = 20
f_sinal = 8
time2 = np.arange(0, 5, 1/Fs_2)
xt_2 = np.cos(2*np.pi*f_sinal*time2)

Fs_1 = 1000
time = np.arange(0, time2[-1], 1/Fs_1)
xt_1 = np.cos(2*np.pi*f_sinal*time)

print('O número de amostras no sinal é {}.'.format(len(time2)))
#print('o vetor xt é {}'.format(xt_2))

plt.figure(figsize = (8,5))
plt.plot(time, xt_1, 'r', linewidth = 1)
plt.stem(time2, xt_2, '-b', label = r"$F_s = {}$ [Hz]".format(Fs_2), basefmt=" ")
plt.legend(loc = 'upper right')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude [-]')
plt.grid(which='both', axis='both')
plt.xlim((0,2))
plt.ylim((-1.6,1.6));
```
