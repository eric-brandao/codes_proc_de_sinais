# Propriedades dos sistemas

```python
# importar as bibliotecas necessárias
import numpy as np # arrays
import matplotlib.pyplot as plt # plots
plt.rcParams.update({'font.size': 14})
from scipy import signal # some signals
import IPython.display as ipd # to play signals
import sounddevice as sd
import soundfile as sf
```

```python
##### x(t) anechoic signal #################
xt, fs = sf.read('voice.wav')
time_xt = np.linspace(0, (len(xt)-1)/fs, len(xt))
##### h(t) anechoic signalimpulse response #################
ht, fs = sf.read('ht.wav')
time_ht = np.linspace(0, (len(ht)-1)/fs, len(ht))
### Convolution
yt = np.convolve(xt,ht[:,0])
yt = 0.95*yt/np.amax(yt)
time_yt = np.linspace(0, (len(yt)-1)/fs, len(yt))
```

## 1. Sem memória

```python
#### y(t)
arrow_y = plt.Line2D((5, 5), (1, 0.5), color = 'r', lw=2.5, marker="",
    markersize=10, markerfacecolor='r')
arrow_y2 = plt.Line2D((5, 5), (0.5, 0.5), color = 'r', lw=2.5, marker="v",
    markersize=10, markerfacecolor='r')

plt.figure()
plt.title(r'$x(t)$')
plt.plot(time_xt, xt, '-k', linewidth = 1, label = 'seno - 1000 [Hz]')

plt.gca().add_line(arrow_y)
plt.gca().add_line(arrow_y2)
#plt.legend(loc = 'best')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude [-]')
plt.xlim((0, time_yt[-1]))
plt.tight_layout()
plt.show()

#### y(t)
arrow_y = plt.Line2D((5, 5), (1, 0.5), color = 'r', lw=2.5, marker="",
    markersize=10, markerfacecolor='r')
arrow_y2 = plt.Line2D((5, 5), (0.5, 0.5), color = 'r', lw=2.5, marker="v",
    markersize=10, markerfacecolor='r')
plt.figure()
plt.title(r'$y(t)$')
plt.plot(time_yt, yt, '-k', linewidth = 1, label = 'seno - 1000 [Hz]')
#plt.legend(loc = 'best')
plt.gca().add_line(arrow_y)
plt.gca().add_line(arrow_y2)
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude [-]')
plt.xlim((0, time_yt[-1]))
plt.tight_layout()
plt.show()
```

```python
# Configurações gerais
fs = 1000
t = np.linspace(0, 1, fs) # vetor temporal
freq = 1 # Frequencia fundamental

xt1 = np.cos(2*np.pi*freq*t)
xt2 = np.cos(2*np.pi*2*freq*t)
xt3 = np.cos(2*np.pi*3*freq*t)

# Figura
plt.figure(figsize=(6,6))
plt.title('Seno')
plt.subplot(3,1,1)
plt.plot(t, xt1, '-b', linewidth = 2, label = '1 [Hz]')
plt.legend(loc = 'lower left')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel(r'$x_1(t)$ [-]')
plt.xlim((0, 1))

plt.subplot(3,1,2)
plt.plot(t, xt2, '-r', linewidth = 2, label = '2 [Hz]')
plt.legend(loc = 'lower left')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel(r'$x_2(t)$ [-]')
plt.xlim((0, 1))

plt.subplot(3,1,3)
plt.plot(t, xt3, '-k', linewidth = 2, label = '3 [Hz]')
plt.legend(loc = 'lower left')
plt.grid(linestyle = '--', which='both')
plt.xlabel('Tempo [s]')
plt.ylabel(r'$x_3(t)$ [-]')
plt.xlim((0, 1))

plt.tight_layout()
plt.show()
```

## 5. Invariância no tempo

```python
# Um impulso unitário
t = np.arange(start = -2, stop = 2, step=0.1) # vetor temporal
xt = np.zeros(len(t))
xt[np.where(np.logical_and(t>=-2e-10,t<=2e-10))] = 1 #complicado assim só por conta de arredondamentos numéricos (as vezes t=0 não é parte do vetor temporal)

plt.figure()
plt.subplot(2,1,1)
plt.stem(t, xt, '-b', label = r'$\delta(t)$', basefmt=" ",
         markerfmt = " ", use_line_collection=  True)
plt.legend(loc = 'upper left')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$x_1(t)$ [-]')
plt.ylim((-0.5, 1.5))
plt.tight_layout()   

# Um impulso unitário
t = np.arange(start = -2, stop = 2, step=0.1)  # vetor temporal
xt = np.zeros(len(t))
xt[np.where(np.logical_and(t>=1-2e-10,t<=1+2e-10))] = 1 # 

plt.subplot(2,1,2)
plt.stem(t, xt, '-r', label = r'$\delta(t-1)$', basefmt=" ",
         markerfmt = " ", use_line_collection=  True)
plt.legend(loc = 'upper left')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$x_2(t)$ [-]')
plt.ylim((-0.5, 1.5))
plt.tight_layout()   
plt.show()
```

```python
fs = 44100
r = np.random.randn(44100)
time = np.linspace(0,3,3*44100)
ht = np.zeros(len(time))
ht[0:fs] = np.exp(-10*time[0:fs])*r

plt.figure()
plt.subplot(2,1,1)
plt.plot(time, ht/np.amax(np.abs(ht)), '-b', label = r'$h(t)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$y_1(t)$ [-]')
plt.xlim((0, 3))
plt.ylim((-1.2, 1.2))

plt.subplot(2,1,2)
plt.plot(time, np.roll(ht/np.amax(np.abs(ht)), fs), '-r', label = r'$h(t-1)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$y_2(t)$ [-]')
plt.xlim((0, 3))
plt.ylim((-1.2, 1.2))
plt.tight_layout()   
plt.show()
```

```python
fs = 44100
r = np.random.randn(44100)
time = np.linspace(0,3,3*44100)
ht = np.zeros(len(time))
ht[0:fs] = np.exp(-10*time[0:fs])*r

ht2 = np.zeros(len(time))
ht2[0:fs] = np.exp(-25*time[0:fs])*r

plt.figure()
plt.subplot(2,1,1)
plt.plot(time, ht/np.amax(np.abs(ht)), '-b', label = r'$h(t)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$y_1(t)$ [-]')
plt.xlim((0, 3))
plt.ylim((-1.2, 1.2))

plt.subplot(2,1,2)
plt.plot(time, 0.7*np.roll(np.flip(ht2)/np.amax(np.abs(ht2)), int(1.3*fs)), '-r', label = r'$h_1(t)$')
plt.legend(loc = 'upper right')
plt.grid(linestyle = '--', which='both')
plt.xlabel('tempo [s]')
plt.ylabel(r'$y_2(t)$ [-]')
plt.xlim((0, 3))
plt.ylim((-1.2, 1.2))
plt.tight_layout()   
plt.show()
```
